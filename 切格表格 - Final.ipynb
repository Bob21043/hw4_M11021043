```python
import cv2
import os

def extract_cells_from_table(image_path, output_directory):
    # 讀取圖像
    image = cv2.imread(image_path)
    
    # 將圖像轉換為灰度
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 使用Canny邊緣檢測
    edges = cv2.Canny(gray, 100, 200)

    # 查找輪廓
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # 確保目標目錄存在
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    # 切割單元格並保存
    for i, contour in enumerate(contours):
        # 計算輪廓的邊界框
        x, y, w, h = cv2.boundingRect(contour)
        
        # 打印單元格的大小
        print(f"Cell {i+1} size: {w} x {h}")

        # 切割單元格
        cell = image[y:y+h, x:x+w]
        
        # 儲存單元格圖像到指定目錄
        cv2.imwrite(os.path.join(output_directory, f"cell_{i+1}.png"), cell)

    # 刪除不符合範圍的單元格
    for filename in os.listdir(output_directory):
        if filename.startswith("cell_"):
            filepath = os.path.join(output_directory, filename)
            img = cv2.imread(filepath)
            if img.shape[0] != 50 or img.shape[1] < 50 or img.shape[1] > 200:
                os.remove(filepath)

# 調用函數
image_path = r'C:\Users\Bob\Unsupervised-Object-Anomaly-Detection-using-Convolutional-Autoencoder-master\T02 LC shift detect\LC shift.png'
output_directory = r'C:\Users\Bob\Unsupervised-Object-Anomaly-Detection-using-Convolutional-Autoencoder-master\T02 LC shift detect\TEST'
extract_cells_from_table(image_path, output_directory)
```
